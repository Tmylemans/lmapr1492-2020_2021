```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd  import XRDCalculator
with MPRester("z18b57ZwfPDtXJVD") as m:

    # Structure for material id : 
    structure = m.get_structure_by_material_id("mp-1170")    
    
    #default values are a calculation using a radiation of CuKa, with the asked wave length
    calculator = XRDCalculator(1.5406, 0.1, None)
    
    #apply on my structure
    diff_pat = calculator.get_pattern(structure)
    
    #show the diffraction pattern
    calculator.show_plot(structure, two_theta_range=(0,30),annotate_peaks=True,ax=None,with_labels=True)
    
    indices = diff_pat.hkls #find the hkl indices
    angles =  diff_pat.x #find the angle 2 theta
    ampli = diff_pat.y #find the scaled intensities
    dist = diff_pat.d_hkls #find the interplanar distance
    
```

```python
print("PREMIER PIC :")
print("Indice de Bravais-Miller hkil : " + str(indices[0][0]['hkl']))
print("Intensité relative = " + str(ampli[0])) 
print("Angle 2Θ = " + str(angles[0]) + "°" )
print("Distance interplanaire = " + str(dist[0]))
```
